# 2. Pie Chart

```python
plt.title("Pie chart of preferred foot of all players in EAFC 24")
gold=df1.loc[df1["overall_rating"]>=75].count()[0]
# gold stores the count of all gold cards
silver=df1.loc[(df1["overall_rating"]<75) & (df1["overall_rating"]>=65)].count()[0]
# Silver holds count of all silver cards
bronze=df1.loc[df1["overall_rating"]<65].count()[0]
#bronze holdscount of all bronze rated cards
plt.pie([gold,silver,bronze],labels=["Gold","Silver","Bronze"],autopct="%0.4f%%",radius=0.85)
# creates a pie chart, based on the count of each rarity int he first attribute
# second attribute is to label each section of the pie
# third attribute is to to put the percentage of that value on the pie
# radius is just to change the size of the circle
plt.legend()
plt.show()
```

# 3. Bar Graph


```python
rating=df1["overall_rating"][df1["overall_rating"]>=75]
# gets all players with above 75 rating
plt.title("Plot of the number of gold players per overall rating")
num=rating.value_counts().sort_index()
# Sorts the ratings in ascending order and takes a count of each rating
plt.bar(num.index,num.values)
# Creates bar graph where x axis is the ratings and y axis is the frequency of ratings
plt.xlabel("Ratings")
plt.ylabel("Number of players with the rating")
plt.xticks(num.index)
# making the scale of x axis 1 unit per rating 
y_max = num.max()  
# Get the maximum value for y-axis scaling
plt.yticks(np.arange(0, y_max + 20, 20)) 
# making the scale for y axis 20 players per unit
plt.show()
```
