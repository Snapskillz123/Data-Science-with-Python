# Sorting Data

```python
df.sort_values("HP")
# Prints HP values in ascending order as that is the default order
```

```python
df.sort_values("HP",ascending=False)
# Prints HP in descending order
```

```python
df.sort_values(["Generation","Attack"])
# Sorts the values first based on the generation, then among that generation it sorts it based on lowest attack to highest attack.
```

```python
df.sort_values(["Generation","Attack"],ascending=[1,0])
# Sorts Generation in ascending order, and in that generation it sorts the pokemons attack in decending order 
```

```python
res_df=df.loc[(df["Type 1"]=="Psychic") & (df["Type 2"]=="Fairy")]
#Prints Psychic type pokemons which are fairy type
res_df=res_df.reset_index()
res_df
# reset_index makes the index values according to this specific condition
```

# Manipulating Data

```python
df["All_Round"]=(df["Attack"]+df["Defense"])/2 
df
# Creating a new column All_Round that is the average of Attack and Defense
```

```python
df= df.drop(columns=["Sp. Def"])
df
# Sp. Def column is being removed from the dataframe. Saving the edited DF in the variable is important
```

```python
df["Sum"]= df.iloc[:, 4:8].sum(axis=1)
df
# Creates a new column Sum that adds columns from HP to Sp. Atk for each row and saves that sum in the new Sums column
# iloc is used to locate index values and return the value of that cell as done earlier
# iloc take arguments[row, column]. Over here ':' selects all rows and 4:8 selects columns of those index values.
# .sum() is used for addition and axis=1 makes sure that the addition in horizontal
```

```python
df.fillna("Mystery")
# Fills null value with "mystery"
```

```python
df = df.replace(to_replace = 'Grass', value ='Nature')
df
#replaces grass type with nature
```
